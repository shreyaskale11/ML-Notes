````python
quantile_lower = 0.01
quantile_higher = 0.99
# count how many rows will be removed if we remove outliers overall for all sensors
df_train_outliers = df_train.copy()
for sensor in range(13):
    sensor_name = f"sensor_{sensor:02d}"
    quantile_lower_value = df_train[sensor_name].quantile(quantile_lower)
    quantile_higher_value = df_train[sensor_name].quantile(quantile_higher)
    df_train_outliers = df_train_outliers.loc[(df_train_outliers[sensor_name] > quantile_lower_value) & (df_train_outliers[sensor_name] < quantile_higher_value)]

print("rows removed: ", df_train.shape[0] - df_train_outliers.shape[0])
print("rows removed %: ", (df_train.shape[0] - df_train_outliers.shape[0])/df_train.shape[0]*100)
print("df_train_outliers shape: ", df_train_outliers.shape)
print("df_outliers without nan shape: ", df_train_outliers.dropna().shape)

# plot distribution for 13 sensors (sensor_00 to sensor_12) after removing outliers
figure = plt.figure(figsize=(16, 8))
for sensor in range(13):
    sensor_name = f"sensor_{sensor:02d}"
    plt.subplot(4, 4, sensor+1)
    plt.hist(df_train[sensor_name], bins=100, range=(df_train[sensor_name].quantile(0.02), df_train[sensor_name].quantile(0.98)))
    plt.title(f"{sensor_name} histogram")
figure.tight_layout(h_pad=1.0, w_pad=0.5)
plt.suptitle('Sensor Histograms Before Outlier Removal', y=1.02)
plt.show()
````

## simple nn model

````python
# simple nn model

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing

# create model
# input layer = 780 , output layer = 2 (0,1) , hidden layer = 3
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[780]),
    layers.Dense(64, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(2)
])

# compile model
model.compile(
    optimizer=tf.optimizers.legacy.Adam(learning_rate=0.01),
    loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

df_train_pivot_X = df_train_pivot.drop(['state','sequence'], axis=1)
df_train_pivot_y = df_train_pivot['state']
print("df_train_pivot_X shape: ", df_train_pivot_X.shape, "df_train_pivot_y shape: ", df_train_pivot_y.shape)
# split train and validation data
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(df_train_pivot_X, df_train_pivot_y, test_size=0.2)

# fit model
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    batch_size=64,
    epochs=63
)

# plot model loss and accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend()
plt.show()

plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.legend()
plt.show()
````

## Lstm

LSTM (Long Short-Term Memory) is a recurrent neural network (RNN) architecture widely used in Deep Learning. It excels at capturing long-term dependencies, making it ideal for sequence prediction tasks.

LSTM is a type of RNN that uses a memory cell, a mechanism that allows it to process sequential data such as text, audio, and video. It is capable of learning long-term dependencies, making it ideal for sequence prediction tasks.

$$ f_t = \sigma_g(W_f \cdot [h_{t-1}, x_t] + b_f) $$
$$ i_t = \sigma_g(W_i \cdot [h_{t-1}, x_t] + b_i) $$
$$ o_t = \sigma_g(W_o \cdot [h_{t-1}, x_t] + b_o) $$
$$ \tilde{C}_t = \sigma_c(W_C \cdot [h_{t-1}, x_t] + b_C) $$
$$ C_t = f_t * C_{t-1} + i_t * \tilde{C}_t $$
$$ h_t = o_t * \sigma_h(C_t) $$

where:
- $x_t$ is the input at time $t$
- $h_t$ is the hidden state at time $t$
- $C_t$ is the cell state at time $t$
- $f_t$ is the forget gate
- $i_t$ is the input gate
- $o_t$ is the output gate
- $\tilde{C}_t$ is the candidate cell state
- $W_f, W_i, W_o, W_C$ are the weight matrices of the gates
- $b_f, b_i, b_o, b_C$ are the bias vectors of the gates
- $\sigma_g$ is the sigmoid function
- $\sigma_c$ is the hyperbolic tangent function
- $\sigma_h$ is the hyperbolic tangent function


````python
# lstm

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing

df_train_pivot_X = df_train_pivot.drop(['state','sequence'], axis=1)
df_train_pivot_y = df_train_pivot['state']

print("df_train_pivot_X shape: ", df_train_pivot_X.shape, "df_train_pivot_y shape: ", df_train_pivot_y.shape)

# split train and validation data
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(df_train_pivot_X, df_train_pivot_y, test_size=0.2)

# reshape train and validation data
X_train = X_train.values.reshape(X_train.shape[0], 60, 13)
X_val = X_val.values.reshape(X_val.shape[0], 60, 13)
print("X_train shape: ", X_train.shape, "X_val shape: ", X_val.shape)

# create model
model = keras.Sequential([
    layers.LSTM(64, input_shape=(60, 13)),
    layers.Dense(64, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(2)
])

# compile model
model.compile(
    optimizer=tf.optimizers.legacy.Adam(learning_rate=0.01),
    loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

# fit model
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    batch_size=64,
    epochs=5
)

# plot model loss and accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend()
plt.show()

plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.legend()
plt.show()
    
````

````python
## 3 lstm stacked layers and 64 units with pytorch with sigmoid activation function 

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset
from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# define dataset class
class SensorDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
    
# Update the model to output a single value
class SensorModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(SensorModel, self).__init__()
        self.num_layers = num_layers
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.lstmlayer1 = nn.LSTM(hidden_size, hidden_size, num_layers, batch_first=True)
        self.lstmlayer2 = nn.LSTM(hidden_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)  # Output a single value
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Set initial hidden and cell states 
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).requires_grad_()
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).requires_grad_()

        # Forward propagate LSTM
        out, _ = self.lstm(x, (h0.detach(), c0.detach()))
        out, _ = self.lstmlayer1(out, (h0.detach(), c0.detach()))
        out, _ = self.lstmlayer2(out, (h0.detach(), c0.detach()))

        # reshape the hidden state of the last time step
        out = out[:, -1, :]
        out = self.fc(out)  # Apply linear layer for 1 output
        out = self.sigmoid(out)
        return out


# Define training function
def train_model(model, train_loader, criterion, optimizer, num_epochs):
    # Train the model
    total_step = len(train_loader)
    loss_list = []
    acc_list = []
    for epoch in range(num_epochs):
        for i, (X_batch, y_batch) in enumerate(train_loader):
            # Convert training data to float
            X_batch = X_batch.float()
            y_batch = y_batch.float()
            
            # Run the forward pass
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            loss_list.append(loss.item())

            # Backprop and perform Adam optimization
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            # Track the accuracy
            total = y_batch.size(0)
            _, predicted = torch.max(outputs.data, 1)
            correct = (predicted == y_batch).sum().item()
            acc_list.append(correct / total)

            if (i + 1) % 100 == 0:
                print('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}, Accuracy: {:.2f}%'
                    .format(epoch + 1, num_epochs, i + 1, total_step, loss.item(),
                        (correct / total) * 100))
    return loss_list, acc_list


# define test function
def test_model(model, test_loader):
    y_pred = []
    y_true = []
    with torch.no_grad():
        correct = 0
        total = 0
        for X_batch, y_batch in test_loader:
            outputs = model(X_batch)
            _, predicted = torch.max(outputs.data, 1)
            y_pred.append(predicted)
            y_true.append(y_batch)
            total += y_batch.size(0)
            correct += (predicted == y_batch).sum().item()

        print('Test Accuracy of the model on the test data: {} %'.format((correct / total) * 100))
    return y_pred, y_true

# pivot test data
df_test_pivot = df_test.pivot(index=['sequence','subject'], columns='step', values=['sensor_00','sensor_01','sensor_02','sensor_03','sensor_04','sensor_05','sensor_06','sensor_07','sensor_08','sensor_09','sensor_10','sensor_11','sensor_12'])

# Flatten the column header
df_test_pivot.columns = [' '.join(map(str, col)).strip() for col in df_test_pivot.columns.values]
# split train data into X and y
df_train_pivot_X = df_train_pivot.drop(['state','sequence'], axis=1)
df_train_pivot_y = df_train_pivot['state']

# split train data into train and validation data
X_train, X_val, y_train, y_val = train_test_split(df_train_pivot_X, df_train_pivot_y, test_size=0.2)
print("X_train shape: ", X_train.shape, "X_val shape: ", X_val.shape)
# reshape train and validation data X_train shape:  (20774, 781) X_val shape:  (5194, 781)
X_train = X_train.values.reshape(X_train.shape[0], 60, 13)
X_val = X_val.values.reshape(X_val.shape[0], 60, 13)
print("X_train shape: ", X_train.shape, "X_val shape: ", X_val.shape)

y_train = y_train.values.reshape(-1, 1)
y_val = y_val.values.reshape(-1, 1)
print("y_train shape: ", y_train.shape, "y_val shape: ", y_val.shape)


# create dataset
train_dataset = SensorDataset(X_train, y_train)
val_dataset = SensorDataset(X_val, y_val)

# create dataloader
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=64, shuffle=True)
val_loader = torch.utils.data.DataLoader(dataset=val_dataset, batch_size=64, shuffle=False)
print("train_loader shape: ", train_loader.dataset.X.shape, "val_loader shape: ", val_loader.dataset.X.shape)
# create model
model = SensorModel(input_size=13, hidden_size=64, num_layers=3, num_classes=1)

# define loss function and optimizer
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
print("model: ", model)
# train model
loss_list, acc_list = train_model(model, train_loader, criterion, optimizer, num_epochs=10)
print("loss_list: ", loss_list, "acc_list: ", acc_list)
# plot loss and accuracy
plt.plot(loss_list, label='train_loss')
plt.legend()
plt.show()

plt.plot(acc_list, label='train_accuracy')
plt.legend()
plt.show()


````

````python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
from transformers import get_linear_schedule_with_warmup

class TrainDataset(torch.utils.data.Dataset):
    def __init__(self, X, y, seq_num):
        super().__init__()
        self.X = X
        self.y = y
        self.seq_num = seq_num
    def __len__(self):
        return len(self.y)
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx[0]//self.seq_num]

class TestDataset(torch.utils.data.Dataset):
    def __init__(self, X, seq_num):
        super().__init__()
        self.X = X
        self.seq_num = seq_num
    def __len__(self):
        return len(self.X) // 60
    def __getitem__(self, idx):
        return self.X[idx]

def prepare_data(data, data_labels, seq_num, data_num, mode='train'):
    if data_labels is not None:
        data_labels = data_labels
    sampler = np.array([list(range(i * seq_num, (i + 1) * seq_num)) for i in range(data_num // seq_num)])
    if mode == 'train':
        dataset = TrainDataset(data, data_labels, seq_num)
    else:
        dataset = TestDataset(data, seq_num)
    dataloader = DataLoader(dataset, batch_size=64, sampler=sampler)
    return dataloader

class LSTM(nn.Module):
    def __init__(self,seq_num = 60,input_dim = 13,lstm_dim = 512,num_layers = 2,num_classes = 1):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, lstm_dim, num_layers, batch_first=True, bidirectional=True)
        self.lstm1 = nn.LSTM(2 * lstm_dim, lstm_dim, num_layers, batch_first=True, bidirectional=True)
        self.lstm2 = nn.LSTM(2 * lstm_dim, lstm_dim, num_layers, batch_first=True, bidirectional=True)
        self.logits = nn.Sequential(
            nn.ReLU(),
            nn.Linear(lstm_dim * seq_num * 2, num_classes),
        )

    def forward(self, x):
        features, _ = self.lstm(x)
        # print("feature 1",features.shape)
        features, _ = self.lstm1(features)
        features, _ = self.lstm2(features)
        features = features.reshape(features.shape[0], -1)
        pred = self.logits(features)
        return pred

def train(epochs,model,optimizer,criterion,scheduler,train_iterator,valid_iterator):  
    for epoch in range(1, epochs + 1):
        training_loss = 0.0
        valid_loss = 0.0

        model.train()
        for batch_idx, batch in enumerate(train_iterator):
            optimizer.zero_grad()
            batch[0] = batch[0].to(device,dtype=torch.float32)
            predict = model(batch[0].float()).squeeze(-1)
            loss = criterion(predict, batch[1].to(device).float())
            loss.backward()
            optimizer.step()
            scheduler.step()
            training_loss += loss.data.item()
        training_loss /= len(train_iterator)

        model.eval()
        for batch_idx, batch in enumerate(valid_iterator):
            batch[0] = batch[0].to(device,dtype=torch.float32)
            predict = model(batch[0].float()).squeeze(-1)
            loss = criterion(predict, batch[1].to(device).float())
            valid_loss += loss.data.item()

        valid_loss /= len(valid_iterator)
        if (epoch % 10 == 1):
            print('Epoch: {}, Training Loss: {:.5f}, '
                 'Validation Loss: {:.5f}'.format(epoch, training_loss, valid_loss))


df_trainX = df_train.drop(['state','subject', 'step','sequence'], axis=1)
df_trainy = df_train['state']
print("df_trainX shape: ", df_trainX.shape, "df_trainy shape: ", df_trainy.shape)

# split train and validation data
X_train, X_val, y_train, y_val = train_test_split(df_trainX, df_trainy, test_size=0.2)
# convert to numpy array
X_train = X_train.values
X_val = X_val.values
y_train = y_train.values
y_val = y_val.values
print("X_train shape: ", X_train.shape, "X_val shape: ", X_val.shape, "y_train shape: ", y_train.shape, "y_val shape: ", y_val.shape)
print("X_train type",type(X_train))
train_dataloader = prepare_data(X_train, y_train, 60, X_train.shape[0])
test_dataloader = prepare_data(X_val, y_val, 60, X_val.shape[0])
print("train_dataloader shape: ", train_dataloader.dataset.X.shape,"test_dataloader shape: ", test_dataloader.dataset.X.shape)


epochs = 1
batch_size = 64
seq_num = 60
if torch.backends.mps.is_available():
    device = torch.device("mps")
elif torch.cuda.is_available():
    device = 'cuda'
else:
    device = 'cpu'
print(f"Using device: {device}")

model = LSTM()
model = model.to(device)
print("model: ", model)
lr = 1e-4
optimizer = torch.optim.Adam(model.parameters(), lr = lr)
num_warmup_steps = int(0.1 * epochs * len(train_dataloader))
num_training_steps = int(epochs * len(train_dataloader))

scheduler = get_linear_schedule_with_warmup(
    optimizer, num_warmup_steps, num_training_steps
)
criterion = nn.MSELoss()

train(epochs, model, optimizer, criterion, scheduler, train_dataloader, test_dataloader)

````

## Singular Value Decomposition (SVD)

SVD is a matrix factorization technique that factorizes a matrix M into the three matrices U, , and V. 

The SVD is commonly written as M = UV^T.

U and V are orthogonal matrices, and  is a diagonal matrix of singular values.

The SVD is computed by default on sparse matrices, as long as they are in CSR or CSC format.

The SVD represents an expansion of the original data in a coordinate system where the covariance matrix is diagonal.

The SVD is commonly used in a wide range of applications including compressing, denoising, and data reduction.

The SVD can be calculated by calling the svd() function.

The function takes a matrix and returns the U,  and V^T elements. The  diagonal matrix is returned as a vector of singular values. The V matrix is returned in a transposed form, e.g. V^T.

The example below defines a 32 matrix and calculates the Singular-value decomposition.

The original matrix is reconstructed from the U, , and V^T elements.

The example then creates a 32 sparse matrix and calculates the Singular-value decomposition.

The original matrix is reconstructed from the U, , and V^T elements.

$$ A = U \Sigma V^T $$




````python
from skimage.color import rgb2gray
from skimage import data
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import svd
 
"""
Singular Value Decomposition
"""
# define a matrix
X = np.array([[3, 3, 2], [2, 3, -2]])
print(X)
# perform SVD
U, singular, V_transpose = svd(X)
# print different components
print("U: ", U)
print("Singular array", singular)
print("V^{T}", V_transpose)
 
"""
Calculate Pseudo inverse
"""
# inverse of singular matrix is just the reciprocal of each element
singular_inv = 1.0 / singular
# create m x n matrix of zeroes and put singular values in it
s_inv = np.zeros(X.shape)
s_inv[0][0] = singular_inv[0]
s_inv[1][1] = singular_inv[1]
print(s_inv)
# calculate pseudo inverse
X_pseudo = V_transpose.T.dot(s_inv.T).dot(U.T)
print("X_pseudo", X_pseudo)
 
""" 
Reconstruct original matrix
"""

# reconstruct original matrix
B = U.dot(s_inv).dot(V_transpose)
print("B", B)
 
"""
Reconstruct original matrix using numpy
"""
# reconstruct original matrix
B = np.linalg.pinv(X)
print("B", B)

````
